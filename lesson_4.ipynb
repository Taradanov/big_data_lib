```python
planets = ['1','2','3','4', '5', '6', '7', '8']
masses = [.005, .18, 1,.1, 315, 95, 15, 17]
```

```python
plt.plot(planets, masses)
plt.show()
```

```python
plt.plot(planets, masses)
plt.yscale(value='log')
plt.show()
```

```python
planet_info = pd.Series(masses, index=planets)
planet_info.plot()
plt.show()
```

```python
planet_info.plot(kind='bar', logy=True)
```

```python
planet_info.plot(kind='barh', logx=True)
plt.show()
```

```python
x = np.linspace(-5,5,101)
x
```

```python
x2=x**2
x2
```

```python
plt.plot(x, x2)
```

```python
x3 = x**3
```

```python
plt.axis([-5,5,-8,8])
plt.plot(x,x2)
plt.plot(x,x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x,x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=.3, hspace=.3)
```
